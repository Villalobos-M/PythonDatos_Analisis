```python
import sqlalchemy as db
from sqlalchemy import text
import pandas as pd
```

```python
engine = db.create_engine("mysql://root:root@172.16.5.4:3310/dw_netflix")
conn = engine.connect()

df_fact = pd.read_sql("SELECT * FROM FactWatchs", con =conn)
df_fact.head()
```

```python
df_user = pd.read_sql("SELECT * FROM dimUser", con =conn)
df_user.head()
```

```python
df_movie = pd.read_sql("SELECT * FROM dimMovie", con =conn)
df_movie.head()
```

Raiting promedio por pelicula

```python
rating_promedio_movie = df_fact.groupby('movieID')['rating'].mean()
rating_promedio_movie
```

```python
rating_prom_movie = df_fact[['movieID', 'rating']].groupby('movieID').mean()
rating_prom_movie
```

```python
rating_prom_movie = rating_prom_movie.reset_index()
rating_prom_movie
```

```python
ratind_user = df_fact[['userID', 'rating']].groupby('userID').mean()
ratind_user = ratind_user.reset_index()
ratind_user
```

Promedio de rating por usuario Premium

```python
df_promedio_premium = df_fact.merge(df_user.query("subscription == 'Premium'"), on="userID").groupby('userID')['rating'].mean() #.mean('rating')
df_promedio_premium =  df_promedio_premium.reset_index()
df_promedio_premium
```
